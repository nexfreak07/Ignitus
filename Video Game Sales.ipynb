This dataset contains a list of video games with sales greater than 100,000 copies. It was generated by a scrape of vgchartz.com.

Fields include

Name - The games name

Platform - Platform of the games release (i.e. PC,PS4, etc.)

Year - Year of the game's release

Genre - Genre of the game

Publisher - Publisher of the game

NA_Sales - Sales in North America (in millions)

EU_Sales - Sales in Europe (in millions)

JP_Sales - Sales in Japan (in millions)

Other_Sales - Sales in the rest of the world (in millions)

Global_Sales - Total worldwide sales.

## Q. What genre and year has the highest video games sales rate being released is recorded?

```python
plt.title('Yearly Video Games Sales')
sns.scatterplot('Year_of_Release','Global_Sales', hue='Genre', s=100,data=data)
```

Answer -: By above Graph we can see that Sports is being rated highestin 2006 approx!

## Q. What is the difference between the sales in 4 different regions namely NA , EU, JP and other regions ?

```python
plt.title('Line Graph for Sales')
plt.plot(data['Year_of_Release'], data['NA_Sales'])
plt.plot(data['Year_of_Release'], data['EU_Sales'])
plt.plot(data['Year_of_Release'], data['JP_Sales'])
plt.plot(data['Year_of_Release'], data['Other_Sales'])
plt.xlabel('Year')
plt.ylabel('Sales')
plt.legend(['NA_Sales',"EU_Sales",'JP_Sales','Other_Sales'])
```

ANSWER: Base on the line graph for sales we can cleary see that the NA region has the most number of gamers followed by EU,JP and Other region respectively. Now, the difference between the sales in different regions is that there are games that has high demands in some regions regions or we could say that the number of gamers in a region affects the rate of video games sales.
